```python
from skimage import io
import os
```

```python
countrylist = []
pathDir =  os.listdir('./')
R = 0
G = 0
B = 0
png = {}

for file in pathDir:
    if file.endswith('png'):
        countrylist.append(file)
 
        
for i in countrylist:
    #print(i)
    f = io.imread(i)
    r = f[:,:,:1].mean()
    g = f[:,:,1:2].mean()
    b = f[:,:,2:3].mean()
    png[i] = [r,g,b]
```
